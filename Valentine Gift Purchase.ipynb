- Quais é o gênero que mais compram?

- A localização influência na hora da compra?
- Renda estimada influência na hora da compra? 
- Qual é a média renda estimada por localização das pessoas que compram?

```python
%%sql

SELECT Purchased, COUNT(*) FROM valentine_gift_purchase
```

```python
%%sql

SELECT gender, COUNT(*) AS Total_generos_compram FROM valentine_gift_purchase
WHERE Purchased = 1
GROUP BY gender
ORDER BY Total_generos_compram DESC;
```

```python
%%sql

SELECT 
    CASE 
        WHEN Age >= 18 AND Age <= 25 THEN  'Faixa etária de 18 a 25'
        WHEN Age >= 26 AND Age <= 35 THEN  'Faixa etária de 26 a 35'
        WHEN Age >= 36 AND Age <= 45 THEN  'Faixa etária de 36 a 45'
        WHEN Age > 45 THEN 'Faixa etária acima de 45'
    END AS faixa_etaria,
    COUNT(*) AS total_pessoas_por_faixa_etaria
FROM valentine_gift_purchase
WHERE Purchased = 1
GROUP BY 
    CASE 
        WHEN Age >= 18 AND Age <= 25 THEN  'Faixa etária de 18 a 25'
        WHEN Age >= 26 AND Age <= 35 THEN  'Faixa etária de 26 a 35'
        WHEN Age >= 36 AND Age <= 45 THEN  'Faixa etária de 36 a 45'
        WHEN Age > 45 THEN  'Faixa etária acima de 45'
    END
   
```

```python
%%sql

SELECT Location, Purchased, COUNT(*) AS Total_de_compras_por_localizacao FROM valentine_gift_purchase
WHERE Purchased = 1
GROUP BY Location, Purchased
ORDER BY Total_de_compras_por_localizacao DESC
```

```python
%%sql

SELECT EstimatedIncome, Purchased, COUNT(*) AS Total_de_compras_por_Renda_estimada FROM valentine_gift_purchase
WHERE Purchased = 1
GROUP BY EstimatedIncome, Purchased
ORDER BY Total_de_compras_por_Renda_estimada DESC
```

```python

%%sql

SELECT Location, ROUND(AVG(EstimatedIncome),2) AS Media_renda_estimada_por_localização  
FROM valentine_gift_purchase
WHERE Purchased = 1
GROUP BY Location
ORDER BY Media_renda_estimada_por_localização DESC
```

```python
# Gera o dataframe

# Portanto, após a execução desse trecho de código, a variável resultado conterá um DataFrame do pandas com os resultados da
# consulta SQL, e as colunas serão rotuladas de acordo com os nomes extraídos da descrição da consulta.
resultado = pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
```

# Conclusão

### Respostas às perguntas de negócio.

- Durante o Valentine's Day, observamos uma tendência marcante em relação ao gênero dos compradores, com o público masculino liderando as compras de presentes. 
- A faixa etária das pessoas que mais compram presentes no Valentine's Day é de 36 a 45 anos. Um fato importante é que a partir dos 26 anos, o volume de compra de presentes aumenta significativamente.
- A localização geográfica também desempenha um papel crucial nas tendências de compra durante o Valentine's Day. As áreas urbanas demonstram um volume considerável de compras em comparação com áreas semi urbanas e rurais. 
- Embora a renda dos consumidores varie, os dados sugerem que não há um padrão claro em relação à renda e às decisões de compra durante o Valentine's Day. 
- Analisando a média de renda por localização de pessoas que compram presente no Valentine's Day, observamos que os consumidores em áreas urbanas têm uma renda média mais alta em comparação com áreas semi urbanas e rurais. locais urbano é de 92062.02, Semi urbano de 70103.45 e Rural de 63477.27.


### Ações
O número de compras realizadas pelo gênero feminino é muito baixo, pois, segundo pesquisas, as mulheres tendem a gastar mais em compras online e em lojas físicas do que os homens. Além disso, as mulheres também gastaram mais por transação, em média. O relatório destacou que as mulheres são responsáveis por cerca de 70% das compras globais. Esta é uma descoberta interessante que pode direcionar estratégias de marketing e produtos para atender às preferências desse grupo específico, já que o potencial de compra das mulheres é maior do que o dos homens, e na análise feita, isso é o oposto, ficando muito aquém em relação às compras dos homens.

Assim, criar websites e melhorar a publicidade para pessoas dessas gerações pode aumentar drasticamente as vendas.
